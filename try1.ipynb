#### 2. Cleaning Scrapped data

```python
def remove_prefix(text, prefix):
    if text.startswith(prefix):
        return text[len(prefix):]
    return text  # or whatever
def removesuffix(self: str, suffix: str, /) -> str:
    # suffix='' should not call self[:-0].
    if suffix and self.endswith(suffix):
        return self[:-len(suffix)]
    else:
        return self[:]
```

```python
for i in range(len(df)):
    df['Fact_Value_2018'][i]=str(df['Fact_Value_2018'][i])
    
```

```python
for i in range(len(df)):
    df['Fact_Value_2019'][i]=str(df['Fact_Value_2019'][i])
```

```python
for i in range(len(df)):
    df['Fact_Value_2018'][i]=remove_prefix( df['Fact_Value_2018'][i],'(')
    df['Fact_Value_2019'][i]=remove_prefix( df['Fact_Value_2019'][i],'(')

```

```python
for i in range(len(df)):
    df['Fact_Value_2018'][i]=removesuffix( df['Fact_Value_2018'][i],')')
    df['Fact_Value_2019'][i]=removesuffix( df['Fact_Value_2019'][i],')')
```

```python
df.head()
```

```python
#delete_spaces
for i in range(len(df)):
    df['Fact_Value_2018'][i]=df['Fact_Value_2018'][i].strip()
    df['Fact_Value_2019'][i]=df['Fact_Value_2019'][i].strip()
```

```python
for i in range(len(df)):
    df['Fact_Value_2018'][i]=df['Fact_Value_2018'][i].replace(",", ".")
    df['Fact_Value_2019'][i]=df['Fact_Value_2019'][i].replace(",", ".")

    
```
